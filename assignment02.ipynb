### import packages for plottion graphs and manipulating data:

```python
import numpy as np
import matplotlib.pyplot as plt
```

### define my function :  $$f(x) = xsinx$$


```python
def myFunction(x) :
    f = np.sin(x) * x
    return f
```

### define the derivative of my function: $$f'(x) = xcosx+sinx$$


```python
def myDerFunction(x) : 
    DF = np.cos(x) * x + np.sin(x)
    return DF
```

### define the domain of the function

```python
x = np.arange(-10,10,0.1)
```

### computing the graph

```python
f = myFunction(x)
DF = myDerFunction(x)
```

### plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.plot(x,f,'b',label="function")
plt.plot(x,DF,'r',label="derivativefunction")
plt.legend(bbox_to_anchor=(1.05, 1), loc =2,borderaxespad=0 )
plt.show()
```

### define the Taylor approximation function : $$\widehat{f} (x) =f'(c)(x-c) + f(c) $$

```python
def myTaylorapprxiFunction1(x) : 
    TF1 = myDerFunction(a) * (x-a) +  myFunction(a)
    return TF1

def myTaylorapprxiFunction2(x) : 
    TF2 = myDerFunction(b) * (x-b) +  myFunction(b)
    return TF2

def myTaylorapprxiFunction3(x) : 
    TF3 = myDerFunction(c) * (x-c) +  myFunction(c)
    return TF3
```

### select the three point that we want

```python
a = -1
b = 0
c = 5
```

### computing the graph

```python
TF1 = myTaylorapprxiFunction1(x)
TF2 = myTaylorapprxiFunction2(x)
TF3 = myTaylorapprxiFunction3(x)
```

### plot the graphs for the Taylor aproximation function with original function and its derivative

```python
plt.figure(1)
plt.plot(x,f,'b',label="function")
plt.plot(x,DF,'r',label="derivativefunction")
plt.plot(x,TF1,'g',label = "Taylorapprxifunction1")
plt.plot(x,TF2,'y',label = "Taylorapprxifunction2")
plt.plot(x,TF3,'',label = "Taylorapprxifunction3")
plt.legend(bbox_to_anchor=(1.05, 1), loc =2,borderaxespad=0 )
plt.show()
```
